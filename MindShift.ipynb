```python
import numpy as mp
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/sample_data/spam.csv')
```

```python
print(df)
```

```python
df['Category'] = df['Category'].map({'spam': 1, 'ham': 0})
```

```python
print(df)
```

```python
x=df['Message']
y=df['Category']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
fe = TfidfVectorizer(min_df=1,stop_words="english",lowercase=True)
xtrainfeature=fe.fit_transform(x_train)
xtestfeature=fe.transform(x_test)
#print(xtrainfeature)
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(xtrainfeature)
```

```python
model=LogisticRegression()
```

```python
model.fit(xtrainfeature,y_train)
```

```python
train_predict=model.predict(xtrainfeature)
train_accuracy=accuracy_score(y_train,train_predict)
print("Training data Accuracy is ",train_accuracy)
```

```python
test_predict=model.predict(xtestfeature)
test_accuracy=accuracy_score(y_test,test_predict)
print("Test data Accuracy is ",test_accuracy)
```

```python
sample_mail=["Had your mobile 11 months or more? U R entitled to Update to the latest colour mobiles with camera for Free! Call The Mobile Update Co FREE on 08002986030"]
samplemail_features=fe.transform(sample_mail)
prediction=model.predict(samplemail_features)
print(prediction)
if(prediction[0]==1):
  print("Spam Mail")
else:
  print("Ham Mail")
```
